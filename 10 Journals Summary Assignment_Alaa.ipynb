
#1 The Application of Improved Decision Tree Algorithm in Data Mining of Employment Rate: Evidence from China

        	The paper has proposed some improvements to ID3 Decision Tree Classifier. The algorithm was used in the data mining of employment rate in China. The data set that was used in testing this improved algorithm was for school students and their employment rate. The proposed improvements overcame two shortcomings of ID3, which are taking information gained as its classification evaluation function to select optimal attribute; however, that is not always true, and ID3 can only deal with discrete values. The proposed improvements can be concluded in two main points. First, adding attribute measure which will be compared with the previous values, and the higher value the measure value, the more information will be provided to classifier. Second, introducing the rate of information gained by dividing information gain by information divided/split, and that solved the ID3 multi-value problem.
 
 
#2 A New Approach for Off-Line Handwritten Arabic
Word Recognition Using KNN Classifier

        	The paper discussed the usage of KNN Classifier for Arabic Handwriting Recognition offline system. Arabic Handwriting is considered to be difficult to recognize as it is similar to some other languages. Hence, KNN Classifier was used for classification and successfully has been tested on IFN/ENIT database. The system discussed three different stages including pre-processing , where text is normalized and filtered, feature extraction to reduce redundancy, and classification using KNN multi-class classification. The recognition rate was affected by the quality of input images of text, and  the effectiveness and pre-processing. The used dataset was taken from IFN/ENIT database, which contains the names of Tunisian towns/ villages postcodes written by more than 1000 people who were asked to write their names as well as filling in some forms related to  their towns/ villages postcodes. To sum up, KNN was used  to classify unlabelled testing set and proved to outperform several existing methods in Arabic Handwritten Arabic word recognition.
 
#3 Application of the Naive Bayes Method to a Decision Support System to provide Discounts (Case Study: PT. Bina Usaha Teknik)

        	This paper discussed the usage of Naive Bayes Classifier in granting discounts using decision support system. It can be said that the system succeeded to provide alternative in calculating price discounts recommendations based on the final probability by any combination of variables/criteria.




#4 MFZ-KNN-A Modified Fuzzy Based K Nearest Neighbour Algorithm
(Case Study: standard UCI data set-Wine)

	A modified fuzzy based KNN algorithm ( MFZ-KNN) was proposed in this paper. The gap that was faced is the complexity of time when classifying data. The dataset that was used in this paper was from UCI data set-Wine (India). Clustering Technique was used to determine fuzzy areas in the dataset as an earlier step of using FCM Classification technique. After applying clustering and classification, test data was classified using FCM and KNN algorithms. In pre-processing stage, fussy clusters were created which were followed by membership matrix creation as a result of FCM algorithm. The proposed improvements showed good results in terms of time complexity as well as accuracy of predictions. 


#5 Inference-Based Naive Bayes: Turning Naive Bayes Cost-Sensitive
(Case Study: UCI data sets and a real-world case study)

    This paper discussed the usage of Inference-Based Naïve Bayes to classify dataset in order to improve prediction accuracy. The technique that was proposed was based on inferencing the order relations of data from the training data and then classifies the instance of test data based on the inferenced order relation that were identified from training dataset. The order relations that were considered are: greater than or equal to, and less than. The relations were identified between an instance’s true classification probability of belonging to the class and the cost-sensitive threshold. The dataset that was used is from UCI datasets. The results showed that the proposed method outperforms other prior existing methods for turning Naïve Bayes cost-sensitive in this particular dataset.

#7 Forecasting Copper Prices By Decision Tree Learning

	This paper proposed an algorithm to predict the copper prices using decision tree classification. The technique used depends on studying the correlation between copper and other metals like gold, silver, and crude oil to determine the features for training. Based on the analysis of the correlation, results showed that those variables are highly correlated with copper, and they were used in decision tree model. Two parameters were passed which are: D= price of other correlated variables, and G= the day after the next day of prediction. The algorithm was tested and succeeded and RMSE was below 5% in average.  

#8 A Feature Dependent Naive Bayes Approach And Its Application To The Software Defect Prediction Problem

	This paper has proposed an algorithm of classification for software defect prediction problem by using Feature Dependent Naive Bayes (FDNB). The dataset was taken from  NASA PROMISE datasets. The dataset contains multiple software modules, and each modules consists of code metrics to be as the classifier input. Line code, McCabe, and Halstead are some of the metrics that were used as input. Insignificant features for predication were filtered ahead. The next step was extracting  the best features using correlation analysis as well as heuristic search algorithm. After that, normalization and discretization were done to improve classifier process. Furthermore FDNB was done to ensure that features independence. Later the results of FDNB classifier performance was compared to Naive Bayes classifier performance and the result of FDNB is slightly perform better. FDNB showed fewer false alarm recorded. One more comparison was done using all data in the pre-processing instead of training set only. The results showed that using all data in pre-processing degrade the performance of the classifier and provide misleading results.

#9 A Naive SVM-KNN Based Stock Market Trend Reversal Analysis For Indian Benchmark Indices

	The paper proposed  stock market trend reversal analysis for Indian Benchmark Indices. The analysis was based on Vector Machine ( SVM ) with K-Nearest Neighbour techniques. The main aim was to forecast the closing price, volatility and momentum of stock market based on dataset input. The used dataset were BSE SENSEX and CNX Nifty( stock exchange in Asia). Firstly, pre-processing was done for time-series data that contains stock opening, stock lower, stock higher and stock closing price for any day. There was a new label added to identify the volatility of the stock. After that,  several stock indicators such as RSI, ATR, Williams, MA were included to be as the input data. There were three steps for the model workflow, which are: updating dataset with pre-processing, dividing datasets into two parts and training was done using SVM classifier, and finally data classification using SVM and K nearest neighbours. The results showed that the  performance of KNN + SVM model was proven to be good, that it is useful not only to forecast closing price, but also to identify volatility and stock momentum.
    
#10 Prediction of Heart Disease Based on Decision Trees

	As decision tree algorithm has been widely used in clinical decision support system, this paper discussed prediction of heart disease using  decision trees classifier. The nature of heart diseases made it hard to be observed with naked eyes and might happen at any time, thus, detecting heart disease using decision support system and decision trees algorithm was proposed. The prediction model was based on patient’s age, gender, chest pain, the resting blood pressure in mmHg, serum cholesterol in mg/d, hereditary, fasting blood pressure in mg/dl, thal and smoking. The model showed adequately successful results for predicting heart diseases occurrence
